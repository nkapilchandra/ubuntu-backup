```python
#BT3051 Assignment 4
#Roll Number: MM14B022
#Collaborators: MM14B005
#Time:

import networkx as nx
import matplotlib.pyplot as plt
import random

def barabasi_albert_graph(G, nNodes, mLinks):
    G_new = G.copy()
    G_new.add_edge(1,2)
    count = 0
    nx.draw(G_new,with_labels=True)
    plt.savefig('initial.png',format='png')
    plt.show()
    
    while (count <= nNodes):
        added = 0
        while (added<mLinks):
            num = G_new.number_of_nodes()
            r = random.randint(1,num)
            deg = G_new.degree(r)
            total_deg = sum(list(dict(G_new.degree(list(range(1,num+1)))).values()))
            prob = deg/total_deg
            threshold = random.random()
            if (prob>threshold):
                G_new.add_edge(r,count+3)
                added += 1
            if (added >= mLinks):
                break

        count += 1
        if (count >= nNodes):
            plot = []
            for i in G_new.degree(list(range(1,num+1))):
                plot = plot + [i]*G_new.degree(list(range(1,num+1)))[i]
            nx.draw(G_new,with_labels=True)
            plt.savefig('final.png',format='png')
            plt.show()
            
#             plt.hist(plot,bins=20)
#             plt.show()
#             print(G_new.number_of_nodes())
            break



if __name__ == '__main__':
    G = nx.Graph()
    G.add_edge(1,2)
    G_new = barabasi_albert_graph(G,98,1)
```
